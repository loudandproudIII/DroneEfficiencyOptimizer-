# Get the thrust and power from speed and RPM

```python
prop = "7x7E"  
V_ms = 30
RPM = 22500

thrust = get_thrust_from_RPM_speed("7x7E",V_ms,RPM)
power  = get_power_from_RPM_speed("7x7E",V_ms,RPM)

print("Thrust [N]:",round(thrust,2),"\n Power [W]:",int(power))
```

# Get power for a know flight condition:

```python
%%time

prop = "7x7E"  
V_ms = 30     # m/s  
Thrust_req = 25 # N  

power = get_power_from_thrust_speed(prop,Thrust_req,V_ms)

print("Power [W]:",int(power))
```
